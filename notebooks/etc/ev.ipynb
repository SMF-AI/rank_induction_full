```python
import numpy as np
import matplotlib.pyplot as plt

# 고유값 분해 예제 행렬 A
A = np.array([[3, 1], [0, 2]])

# 고유값 및 고유벡터 계산
eigenvalues, eigenvectors = np.linalg.eig(A)
print(eigenvectors)
# 그리드 생성
x = np.linspace(-2, 2, 10)
y = np.linspace(-2, 2, 10)
X, Y = np.meshgrid(x, y)
grid = np.vstack([X.ravel(), Y.ravel()])

# 변환 후 그리드
transformed_grid = A @ grid

# 원래의 단위벡터 및 변환된 벡터
unit_vectors = np.eye(2)  # [[1,0], [0,1]]
transformed_vectors = A @ unit_vectors

# 시각화
import matplotlib.pyplot as plt

# 임의의 벡터 두 개 정의
v1 = np.array([1, 2])
v2 = np.array([-1, 1])

# 변환 후 벡터
v1_transformed = A @ v1
v2_transformed = A @ v2

# x, y축에 평행한 격자형 그리드 생성
x = np.linspace(-3, 3, 10)
y = np.linspace(-3, 3, 10)
X, Y = np.meshgrid(x, y)
grid = np.vstack([X.ravel(), Y.ravel()])

# 변환 후 그리드
transformed_grid = A @ grid

# Figure 및 Subplots 설정
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# 1. 원본 벡터 및 격자형 그리드 표시 (좌측 그래프)
axs[0].plot(X, Y, color='gray', alpha=0.5)  # 수직선 그리드
axs[0].plot(X.T, Y.T, color='gray', alpha=0.5)  # 수평선 그리드
axs[0].quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color="blue", label="Vector 1")  # 벡터 1
axs[0].quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color="green", label="Vector 2")  # 벡터 2

axs[0].set_xlim(-3, 3)
axs[0].set_ylim(-3, 3)
axs[0].axhline(0, color='black', linewidth=0.5)
axs[0].axvline(0, color='black', linewidth=0.5)
axs[0].set_title("Original Grid and Vectors")
axs[0].legend()
axs[0].set_aspect('equal')

# 2. 변환 후 벡터 및 변환된 격자 표시 (우측 그래프)
transformed_X = (A @ np.vstack([X.ravel(), Y.ravel()])).reshape(2, *X.shape)
transformed_Y = (A @ np.vstack([X.T.ravel(), Y.T.ravel()])).reshape(2, *Y.shape)

axs[1].plot(transformed_X[0], transformed_X[1], color='gray', alpha=0.5)  # 변환 후 수직선
axs[1].plot(transformed_Y[0], transformed_Y[1], color='gray', alpha=0.5)  # 변환 후 수평선
axs[1].quiver(0, 0, v1_transformed[0], v1_transformed[1], angles='xy', scale_units='xy', scale=1, color="blue", label="Transformed Vector 1")  # 변환 후 벡터 1
axs[1].quiver(0, 0, v2_transformed[0], v2_transformed[1], angles='xy', scale_units='xy', scale=1, color="green", label="Transformed Vector 2")  # 변환 후 벡터 2

# 고유벡터 표시
for i in range(len(eigenvalues)):
    ev = eigenvectors[:, i] * eigenvalues[i]  # 고유벡터 방향으로 고유값만큼 변환
    axs[1].quiver(0, 0, ev[0], ev[1], angles='xy', scale_units='xy', scale=1, color="red", label=f"Eigenvector {i+1}")

axs[1].set_xlim(-5, 5)
axs[1].set_ylim(-5, 5)
axs[1].axhline(0, color='black', linewidth=0.5)
axs[1].axvline(0, color='black', linewidth=0.5)
axs[1].set_title("Transformed Grid, Vectors, and Eigenvectors")
axs[1].legend()
axs[1].set_aspect('equal')

plt.show()

```
